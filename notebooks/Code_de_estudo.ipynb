# Biblioteca Beautiful Soup - Prática I

```python
import pandas as pd
import requests

from datetime import datetime
from bs4 import BeautifulSoup


# buscar  a url
url = 'https://www2.hm.com/en_us/men/products/jeans.html'

# headers é um local padrão para que não tenha problema em coletar dados
# fazer uma requisição para que a url funcione
headers = {'User-Agent' : 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'}

page = requests.get(url, headers=headers)

# instaciar a classe principal e parser (html ou xlm = tem que instlar)
# organiza o html
soup = BeautifulSoup(page.text, 'html.parser')

# usando find para retorar so o elemento na lista toda como find_all
products = soup.find('ul', class_='products-listing small')

# criar uma lista com todos os produtos
products_list = soup.find_all('article', class_='hm-product-item')

"""
olhe qual o tipo se retornar 'bs4.element.Tag' pode use o metodo get para agilizar
type(productsu_list[0])
o get você so precisa pegar a variavel que vc quer que retorne
productsu_list[0].get('data-articlecode')
"""

# criar uma lista com todos os ID's dos produtos
# cria um loop for mais enchuto EX: product_id = [p.get('data-articlecode') for p in productsu_list]

# product_id
product_id = [p.get('data-articlecode') for p in products_list]

#product_category
product_category = [p.get('data-category') for p in products_list]


# criar lista para selecionar os nomes
# selecione so o nome doproduto 'products_listb[8].get_text()' usa para ver se deu certo
products_listb = products.find_all('a', class_='link')

# loop for enchuto, ja fez o teste anterior pode usar o .get_text()

# product_name
product_name = [p.get_text() for p in products_listb]

#criar lista com os preços
products_listc = products.find_all('span', class_='price regular')

# analiza se esta certo 'products_listc[0].get_text()'

# product_price
product_price = [p.get_text() for p in products_listc]

# product_color

# product_composition

# montar um dataframe com a coleta de dados
data = pd.DataFrame([product_id, product_category, product_name, product_price]).T # o .T é para virar o dataframe

# colocar nome nas colunas
data.columns = ['product_id', 'product_category', 'product_name', 'product_price']

# vamos colocar a data para que  no futuro possa fazer comparações
#scrapy datetime
data['scrapy_datetime'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

data.head()
```

```python
data.dtypes
```

```python
# limpar product_price tirar o $
data['product_price'] = data['product_price'].apply(lambda x: x.replace('$', '') if pd.notnull(x) else x)
data.head()
```

# Biblioteca Beautiful Soup - Prática III

### One Product

```python
import pandas as pd
import requests

from datetime import datetime
from bs4 import BeautifulSoup


# API requests
# buscar  a url(esse url e so da pagina de uma calça para achar a cor e a composição)
url = 'https://www2.hm.com/en_us/productpage.0985159002.html'

# headers é um local padrão para que não tenha problema em coletar dados
# fazer uma requisição para que a url funcione
headers = {'User-Agent' : 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'}

page = requests.get(url, headers=headers)

# Beautiful soup object
# instaciar a classe principal e parser (html ou xlm = tem que instlar)
# organiza o html
soup = BeautifulSoup(page.text, 'html.parser')

# ==================== product_color ======================
# olha se tem todas a informações necessarias use: soup.find_all('li', class_='list-item')
# achar a cor e ver se retorna ela: soup.find_all('a', class_='filter-option miniature')[0].get('data-color')
product_list = soup.find_all('a', class_='filter-option miniature')

# loop for 
color_name = [p.get('data-color') for p in product_list]

# product_id
product_id = [p.get('data-articlecode') for p in product_list]

# cria o dataframe
df_color = pd.DataFrame([product_id, color_name]).T # o .T é apara mudar a direção das colunas
df_color.columns = ['product_id', 'color_name']

# generate style_id + color_id
# criar as colunas novas [style_id, color_id]
# os tres ultimos numeros do 'product_id' ou o id são denominada as cores e o resto e resto e o codigo do produto

df_color['style_id'] = df_color['product_id'].apply(lambda x: x[:-3])
df_color['color_id'] = df_color['product_id'].apply(lambda x: x[-3:])

# ==================== composition  ======================
# pode ser usado o get_text() no lugar do get para pegar so o texto
# use o .split('\n') para para remover os \n
# limpar as posições vacias com o metodo filter e retorne como lista se não resolve
# EX:
# list(filter(None, soup.find_all('div', class_='pdp-description-list-item')[1].get_text().split('\n')))

product_composition_list = soup.find_all('div', class_='pdp-description-list-item')

# loop for
product_composition = [list(filter(None, p.get_text().split('\n'))) for p in product_composition_list]

# criar o dataframe e localizar a primeira linha
df_composition = pd.DataFrame(product_composition).T
df_composition.columns = df_composition.iloc[0]

# deletar a primeira linha do dataframe
# [1:]vai retorna a priemira linha depois do 0 e as demais abixo
# o metodo fillna(method='ffill') se utiliza quando vc quer repetir dados de cima para baixo
df_composition = df_composition.iloc[1:].fillna(method='ffill')

# generate style_id + color_id
# criar as colunas novas [style_id, color_id]
# os tres ultimos numeros do 'Art.No' ou o id são denominada as cores e o resto e resto e o codigo do produto

df_composition['style_id'] = df_composition['Art. No.'].apply(lambda x: x[:-3])
df_composition['color_id'] = df_composition['Art. No.'].apply(lambda x: x[-3:])

# juntar os dataframes com marge
data_sku = pd.merge(df_color, df_composition[['style_id', 'Fit', 'Composition']], how='left', on='style_id')
```

```python
import pandas as pd
import requests

from datetime import datetime
from bs4 import BeautifulSoup

# O HEADERS NÃO PRECISA ENTRAR NO LOOP FOR PQ SO PRECISA LER UMA VEZ
# headers é um local padrão para que não tenha problema em coletar dados
# fazer uma requisição para que a url funcione
headers = {'User-Agent' : 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'}

# CRIAR UMA VARIAVEL PARA RECEBER O RESULTADO DO DATAFRAME
df_details = pd.DataFrame()

# cria uma variavel que vai receber uma lista de colunas que são os resultados a mais
# unique columns for all products
aux = []
# depois que rodar o aux vc usa o set(aux) para ver as colunas sem se repetir
# retire as colunas que vc use ou que vc criou 
cols = ['Art. No.', 'Composition', 'Fit', 'Size']
# crie um dataframe para receber essas colunas acima
df_pattern = pd.DataFrame(columns=cols)


# API requests

# AGORA TEM QUE USARA URL PARA PECORRER TODOS OS PRODUTOS
# usar o for para isso

for i in range(len(data)):
    # buscar  a url
    url = 'https://www2.hm.com/en_us/productpage.' + data.loc[i, 'product_id'] + '.html'

    page = requests.get(url, headers=headers)

    # Beautiful soup object
    # instaciar a classe principal e parser (html ou xlm = tem que instlar)
    # organiza o html
    soup = BeautifulSoup(page.text, 'html.parser')

    # ==================== product_color ======================
    # olha se tem todas a informações necessarias use: soup.find_all('li', class_='list-item')
    # achar a cor e ver se retorna ela: soup.find_all('a', class_='filter-option miniature')[0].get('data-color')
    product_list = soup.find_all('a', class_='filter-option miniature')

    # loop for 
    color_name = [p.get('data-color') for p in product_list]

    # product_id
    product_id = [p.get('data-articlecode') for p in product_list]

    # cria o dataframe
    df_color = pd.DataFrame([product_id, color_name]).T # o .T é apara mudar a direção das colunas
    df_color.columns = ['product_id', 'color_name']

    # generate style_id + color_id
    # criar as colunas novas [style_id, color_id]
    # os tres ultimos numeros do 'product_id' ou o id são denominada as cores e o resto e resto e o codigo do produto

    df_color['style_id'] = df_color['product_id'].apply(lambda x: x[:-3])
    df_color['color_id'] = df_color['product_id'].apply(lambda x: x[-3:])

    # ==================== composition  ======================
    # pode ser usado o get_text() no lugar do get para pegar so o texto
    # use o .split('\n') para para remover os \n
    # limpar as posições vacias com o metodo filter e retorne como lista se não resolve
    # EX:
    # list(filter(None, soup.find_all('div', class_='pdp-description-list-item')[1].get_text().split('\n')))

    product_composition_list = soup.find_all('div', class_='pdp-description-list-item')

    # loop for
    product_composition = [list(filter(None, p.get_text().split('\n'))) for p in product_composition_list]

    # criar o dataframe e localizar a primeira linha
    df_composition = pd.DataFrame(product_composition).T
    df_composition.columns = df_composition.iloc[0]

    # deletar a primeira linha do dataframe
    # [1:]vai retorna a priemira linha depois do 0 e as demais abixo
    # o metodo fillna(method='ffill') se utiliza quando vc quer repetir dados de cima para baixo
    df_composition = df_composition.iloc[1:].fillna(method='ffill')
    
    # concatenar as colunas do df_composition com df_pattern 
    # assim retornando para o df_composition
    df_composition = pd.concat([df_pattern, df_composition], axis=0)

    # generate style_id + color_id  
    # criar as colunas novas [style_id, color_id]
    # os tres ultimos numeros do 'Art.No' ou o id são denominada as cores e o resto e resto e o codigo do produto

    df_composition['style_id'] = df_composition['Art. No.'].apply(lambda x: x[:-3])
    df_composition['color_id'] = df_composition['Art. No.'].apply(lambda x: x[-3:])

    # para coloar os outros elementos estão dentro e não aparecem é so criar uma variavel com as colunas
    aux = aux + df_composition.columns.tolist()
    
    
    # juntar os dataframes com marge
    data_sku = pd.merge(df_color, df_composition[['style_id', 'Fit', 'Composition', 'Size']], how='left', on='style_id')
    
    # concatenar todos os resultados e concatenar com a variavel para que vai receber
    df_details = pd.concat([df_details, data_sku], axis=0)
    
    
    
# Join showroom data + df_details
# commo o data e nossa vitrine principal vamo fazer algumas coluas e juntar
data['style_id'] = data['product_id'].apply(lambda x: x[:-3])
data['color_id'] = data['product_id'].apply(lambda x: x[-3:])

# juntar datafremes
data_raw = pd.merge(data, df_details[['style_id', 'color_name', 'Fit', 'Composition', 'Size']], how='left', on='style_id')
data_raw
```
